```python
import pandas as pd
import numpy as np
import re
import matplotlib.pylab as plt
import seaborn as sns
import gc
%matplotlib inline
```

```python
filedir = '/home/hui/文档/1113/20171110.dat'
data = pd.read_csv(filedir,header=None,index_col=0,names=['ticker','bid_price','bid_volume','ask_price','ask_volume','last_price','last_volume','open_interest','turnover'])
data.sort_index(inplace=True)
```

```python
def dropMs(df):
    lst = list(df.index.values)
    res = []
    for time in lst:
        s = re.split(r'[:.]', time)
        s = s[:-1]
        if s[-1] == '60':
            s[-1] = '00'
            s[-2] = str(int(s[-2]) + 1)
            if s[-2] == '61':
                s[-2] = '01'
                s[-1] = str(int(s[-1]) + 1)
        if s[-2] == '60':
                s[-2] = '00'
                s[-1] = str(int(s[-1]) + 1)
        s = s[0] + ':' + s[1] + ':' + s[2]
        pd.to_datetime(s, format='%H:%M:%S')
        res.append(s)
    df.index = pd.DatetimeIndex(res)

def reSample(df,freq = '5T', how = 'first'):
    df = df.resample(freq, how)

    
def timeNormalize(df):   #处理交易时间不规整
    lst = list(df.index.values)
    res = []
    for time in lst:
        s = re.split(r'[:.]',time)
        if int(s[-1]) <= 500:
            s = s[0]+':'+s[1]+':'+s[2]+'.'+'500'
        elif int(s[-1]) < 1000:
            s[-2] = str(int(s[-2]) + 1)
            if int(s[-2]) == 60:
                s[-3] = str(int(s[-3]) + 1)
                s[-2] = '00'
            elif len(s[-2]) == 1:
                s[-2] = '0' + s[-2]
            s = s[0] + ':'+ s[1]+':'+s[2]+'.'+'000'
        res.append(s)
    df.index = res

    
def midPrice(df):  #计算mid_pricr
    df.loc[:,'mid_price'] = (df.loc[:,'bid_price'] + df.loc[:,'ask_price']) / 2

def midRet(df):
    res = [0]
    for i in range(1,df.shape[0]):
        if df.mid_price.values[i-1] == 0:
            temp = 0
        else:
            temp = (df.mid_price.values[i] - df.mid_price.values[i-1])/df.mid_price.values[i-1]
        res.append(temp)
    df.loc[:,'mid_price_return'] = res
        

def midRetCum(df):
    df.loc[:,'mid_price_return_cum'] = df.loc[:,'mid_price_return'].values.cumsum()


def calcAll(df):
    midPrice(df)
    midRet(df)
    midRetCum(df)
    
def corrMat(df,lst = None, returnType = 'corrmat',threshold = 1000, keywd = 'mid_price_return_cum',plot = False, size = (12,7)): #计算相关系数矩阵
    #默认计算所有ticker间的相关系数，返回相关系数矩阵，可通过修改lst变为计算lst内元素的相关系数，通过修改returnType调整返回align后的DataFrame
    '''if lst == None then calculate all elements otherwise only calculate elements in lst;
       if returnType == corrmat then return corr Matrix otherwise return aligned DataFrame;'''
    if lst:
        lst = filterName(lst)
    else:
        lst = list(set(df.ticker))
        lst = filterName(lst)
    lst.sort()
    res = pd.DataFrame()
    elem = lst[0]
    temp = df[df['ticker'] == elem]
    calcAll(temp)
    temp = temp.rename(columns = {keywd:elem})
    res[elem] = temp[elem]
    for elem in lst[1:]:
        temp = df[df['ticker'] == elem]
        if temp.shape[0] < threshold:
            continue
        else:
            calcAll(temp)
            temp = temp.rename(columns = {keywd:elem})
            res, temp = res.align(temp,join = 'outer')
            res = pd.concat([res.iloc[:,:-1], temp[elem]],axis = 1)
            res = res.fillna(method='ffill')
            res = res.fillna(method='bfill')
            res = res.dropna(axis = 1,how='all')
            res['ind'] = res.index
            res = res.drop_duplicates(subset='ind')
#         print res.shape
    if plot:
        plt.subplots(figsize = size)
        sns.set(font_scale=1.25)
        sns.heatmap(res.corr(), cbar=True, annot=True,square=True, fmt='.3f', annot_kws={'size': 10})
        plt.xticks(rotation = 90)
        plt.yticks(rotation = 0)
    elif returnType == 'corrmat':
        return res.corr()
    else:
        return res

    
def filterName(lst):  #判断是否为期权
    '''judge option or not'''
    ans = []
    for name in lst:
        if not ('-P-' in name or '-C-' in name or 'SR' in name):
            ans.append(name)
    return ans
    
def corrNormalize(mat):   #相关系数矩阵减单位矩阵
    size = mat.shape[0]
    eye_mat = np.eye(size)
    mat = mat - eye_mat
    return mat
```

```python
def getLength(df):
    dict = {}
    lst = list(set(df['ticker'].values))
    for name in lst:
        if not filterName(name):
            temp = df[df['ticker'] == name]
            if temp.shape[0] >= 1000:
                dict[name] = temp.shape[0]
    return dict
```

```python
def findMostInType1(df):  
    lst = list(set(df.ticker))
    existed = []
    length = {}
    most = {}
    for name in lst:
        temp = df[df['ticker'] == name]
        l = temp.shape[0]
        if name[:2] in existed:
            if l > length[name[:2]]:
                most[name[:2]] = name
                length[name[:2]] = l
        else:
            existed.append(name[:2])
            length[name[:2]] = l
            most[name[:2]] = name
        del temp; del l; gc.collect()
    return most

def findMostInType(df):
    dic = df.groupby('ticker')['turnover'].max()
    lst = dic.index.values
    existed = []
    length = {}
    most = {}
    for name in lst:
        l = dic[name]
        if name[:2] in existed:
            if l > length[name[:2]]:
                most[name[:2]] = name
                length[name[:2]] = l
        else:
            existed.append(name[:2])
            length[name[:2]] = l
            most[name[:2]] = name
    return most

def singleType(s,df, plot = True, size = (12,7)):   #计算单一品种不同ticker间的相关系数
    ansLst = []
    for name in list(set(df.ticker)):
        if s in name:
            ansLst.append(name)
    res = corrMat(data,ansLst)
    if plot:
        plt.subplots(figsize = size)
        sns.set(font_scale=1.25)
        sns.heatmap(res, cbar=True, annot=True,square=True, fmt='.3f', annot_kws={'size': 10})
        plt.xticks(rotation = 90)
        plt.yticks(rotation = 0)
    return res

def findNstElem(df, ticker, k = 10):
    #correlation to specified ticker
    cols = df.corr().nlargest(k, ticker)[ticker].index
    cm = np.corrcoef(df[cols].values.T)
    sns.set(font_scale=1.25)
    hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
    plt.show()
```
